```python
import pandas as pd
import matplotlib.pyplot as plt

# Sample data
data = {
    'Month': ['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun',
              'Jul', 'Aug', 'Sep', 'Oct', 'Nov', 'Dec'],
    'Electricity_kWh': [3200, 2800, 3000, 2700, 2600, 3100,
                        4000, 4200, 3900, 3300, 3100, 3400],
    'Gas_kWh': [1500, 1400, 1200, 900, 500, 300,
                200, 250, 400, 800, 1200, 1400]
}

# Create DataFrame
df = pd.DataFrame(data)
df['Total_Energy_kWh'] = df['Electricity_kWh'] + df['Gas_kWh']

# Plot 1: Electricity
plt.figure(figsize=(10, 5))
plt.plot(df['Month'], df['Electricity_kWh'], color='blue', marker='o')
plt.title('Monthly Electricity Usage')
plt.xlabel('Month')
plt.ylabel('Electricity (kWh)')
plt.grid(True)
plt.tight_layout()
plt.show()

# Plot 2: Gas
plt.figure(figsize=(10, 5))
plt.plot(df['Month'], df['Gas_kWh'], color='green', marker='o')
plt.title('Monthly Gas Usage')
plt.xlabel('Month')
plt.ylabel('Gas (kWh)')
plt.grid(True)
plt.tight_layout()
plt.show()

# Plot 3: Total Energy
plt.figure(figsize=(10, 5))
plt.plot(df['Month'], df['Total_Energy_kWh'], color='black', linestyle='--', marker='o')
plt.title('Monthly Total Energy Consumption')
plt.xlabel('Month')
plt.ylabel('Total Energy (kWh)')
plt.grid(True)
plt.tight_layout()
plt.show()
```
